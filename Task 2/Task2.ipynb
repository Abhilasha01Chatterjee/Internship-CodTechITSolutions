```python
df = df.drop(columns = ['Education Level','Diabetes','Cholesterol Level','Dietary Habits','Air Pollution Exposure','Employment Status','Marital Status','Social Engagement Level','Income Level','Urban vs Rural Living'])
# to drop the unwanted columns from the dataset
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
# to convert all the string values into numeric terms
df['Country'] = encoder.fit_transform(df['Country'])
df['Gender'] = encoder.fit_transform(df['Gender'])
df['Physical Activity Level'] = encoder.fit_transform(df['Physical Activity Level'])
df['Smoking Status'] = encoder.fit_transform(df['Smoking Status'])
df['Alcohol Consumption'] = encoder.fit_transform(df['Alcohol Consumption'])
df['Hypertension'] = encoder.fit_transform(df['Hypertension'])
df['Family History of Alzheimers'] = encoder.fit_transform(df['Family History of Alzheimers'])
df['Depression Level'] = encoder.fit_transform(df['Depression Level'])
df['Sleep Quality'] = encoder.fit_transform(df['Sleep Quality'])
df['Genetic Risk Factor'] = encoder.fit_transform(df['Genetic Risk Factor'])
df['Stress Levels'] = encoder.fit_transform(df['Stress Levels'])
df['Alzheimers Diagnosis'] = encoder.fit_transform(df['Alzheimers Diagnosis'])
```

### Splitting data into training and testing sets

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)
```

```python
X_train
```

```python
y_train
```

## Building the models

### 1) Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
regressor_lr = LinearRegression()
```

```python
regressor_lr.fit(X_test, y_test)
```

```python
y_pred = regressor_lr.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

### 2) Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor()
```

```python
regressor_rf.fit(X_train, y_train)
```

```python
y_pred = regressor_rf.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```
